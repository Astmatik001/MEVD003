```python
data = df[['playlist_genre', 'track_popularity']].dropna()

genres = data['playlist_genre'].unique()
popularity_by_genre = [
    data[data['playlist_genre'] == genre]['track_popularity']
    for genre in genres
]

plt.figure(figsize=(10, 6))
plt.boxplot(popularity_by_genre, tick_labels=genres, showfliers=False)
plt.xticks(rotation=45)
plt.ylabel('Track Popularity')
plt.xlabel('Playlist Genre')
plt.title('Genre vs Popularity')
plt.tight_layout()
plt.show()
```

Гипотеза №1 - Средняя популярность треков отличается между жанрами.

Гипотеза подтверждается графиком.

Значимость результата ниже среднего - представление о данных изменилось не сильно.
Вполне ожидаемо, поп - самая популярная категория, а джазз - самая непопулярная.
Не ожидал большую популярность классической музыки.

```python
data = df[['danceability', 'track_popularity']].dropna()

plt.figure(figsize=(8, 4))

plt.scatter(
    data['danceability'],
    data['track_popularity'],
    alpha=0.4
)
plt.xlabel('Danceability')
plt.ylabel('Track Popularity')
plt.title('Danceability vs Popularity')

plt.tight_layout()
plt.show()
```
